### 📌 Objective
Analyze Netflix's content library to uncover:
- Content type distribution (Movies vs TV Shows)
- Top and bottom contributing directors
- Country-wise content production
- Rating categories and duration preferences
- Yearly and monthly release trends


## 1. 📦 Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns
from IPython.display import display

# Plot styling
sns.set_theme(style='whitegrid', palette='muted')
plt.rcParams['figure.dpi'] = 100
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['axes.labelsize'] = 12
```

## 2. 📂 Load Dataset

```python
# Load the Netflix dataset
# Update the path below to point to your local CSV file
df = pd.read_csv('netflix_data.csv')

print(f"Dataset shape: {df.shape}")
df.head()
```

## 3. 🔍 Data Exploration

```python
# Dataset structure and data types
df.info()
```

```python
# Check for missing values
missing = df.isnull().sum()
print("Missing values per column:")
print(missing[missing > 0])
```

```python
# Statistical summary
df.describe()
```

## 4. 📊 Analysis & Visualizations

### 4.1 Movies vs TV Shows

```python
fig, ax = plt.subplots(figsize=(7, 5))
sns.countplot(x=df['Category'], hue=df['Category'], ax=ax, palette='Set2', legend=False)
ax.set_title('Content Type Distribution on Netflix')
ax.set_xlabel('Category')
ax.set_ylabel('Count')
plt.tight_layout()
plt.show()
```

> **Insight:** **Movies significantly outnumber TV Shows** on Netflix, making up the majority of the content library.

### 4.2 Top 8 Directors by Show Count

```python
director_counts = df['Director'].value_counts()

fig, ax = plt.subplots(figsize=(10, 6))
director_counts.nlargest(8).plot(kind='bar', ax=ax,
    color=['#1f77b4','#ff7f0e','#2ca02c','#d62728','#9467bd','#8c564b','#e377c2','#7f7f7f'])
ax.set_title('Top 8 Directors by Number of Shows on Netflix')
ax.set_xlabel('Director')
ax.set_ylabel('Number of Shows')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

### 4.3 Bottom 8 Directors by Show Count

```python
fig, ax = plt.subplots(figsize=(10, 6))
director_counts.nsmallest(8).plot(kind='bar', ax=ax,
    color=['#1f77b4','#ff7f0e','#2ca02c','#d62728','#9467bd','#8c564b','#e377c2','#7f7f7f'])
ax.set_title('Bottom 8 Directors by Number of Shows on Netflix')
ax.set_xlabel('Director')
ax.set_ylabel('Number of Shows')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

### 4.4 Top 10 Countries by Content Volume

```python
country_counts = df['Country'].value_counts().nlargest(10)

fig, ax = plt.subplots(figsize=(10, 5))
country_counts.plot(kind='line', marker='o', ax=ax, color='steelblue', linewidth=2)
ax.set_title('Top 10 Countries with the Most Netflix Shows')
ax.set_xlabel('Country')
ax.set_ylabel('Number of Shows')
plt.xticks(rotation=20, ha='right')
plt.tight_layout()
plt.show()
```

> **Insight:** The **United States** produces by far the highest number of Netflix shows, followed by India and the UK.

### 4.5 Content Rating Distribution

```python
fig, ax = plt.subplots(figsize=(12, 5))
sns.countplot(x=df['Rating'], ax=ax, color='steelblue',
              order=df['Rating'].value_counts().index)
ax.set_title('Distribution of Content Ratings on Netflix')
ax.set_xlabel('Rating')
ax.set_ylabel('Count')
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```

> **Insight:** **TV-MA** (Mature Audiences) is the most common content rating on Netflix, indicating the platform skews toward adult content.

### 4.6 Top 15 Show Durations

```python
duration_counts = df['Duration'].value_counts()

fig, ax = plt.subplots(figsize=(10, 6))
duration_counts.nlargest(15).plot(kind='line', marker='o', ax=ax, color='green', linewidth=2)
ax.set_title('Top 15 Most Common Show Durations')
ax.set_xlabel('Duration')
ax.set_ylabel('Number of Shows')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

> **Insight:** **1 Season** is the most common duration for TV shows, suggesting many series are limited or mini-series.

### 4.7 Top 50 Content Types

```python
type_counts = df['Type'].value_counts().nlargest(50)
top_types_df = type_counts.reset_index()
top_types_df.columns = ['Type', 'Count']

styled = top_types_df.style.background_gradient(cmap='Blues').set_caption('Top 50 Netflix Content Types')
display(styled)
```

> **Insight:** **Documentaries** are the most prevalent content type in the Netflix catalog, reflecting strong investment in non-fiction storytelling.

### 4.8 Yearly Release Trend

```python
df['Release_Date'] = pd.to_datetime(df['Release_Date'], errors='coerce')
df['release_year'] = df['Release_Date'].dt.year

yearly = df.groupby('release_year').size().reset_index(name='count')

fig, ax = plt.subplots(figsize=(12, 6))
colors = cm.Blues(yearly['count'] / yearly['count'].max())
ax.bar(yearly['release_year'], yearly['count'], color=colors)
ax.set_title('Number of Shows Released per Year on Netflix')
ax.set_xlabel('Release Year')
ax.set_ylabel('Number of Shows')
plt.tight_layout()
plt.show()
```

> **Insight:** Content releases peaked between **2018 and 2020**, with **2019** being the single highest year for new additions.

### 4.9 Top 10 Release Months

```python
df['release_month'] = df['Release_Date'].dt.strftime('%B %Y')
monthly = df.groupby('release_month').size().reset_index(name='count')
top_months = monthly.nlargest(10, 'count')

fig, ax = plt.subplots(figsize=(14, 6))
colors = cm.Blues(top_months['count'] / top_months['count'].max())
ax.bar(top_months['release_month'], top_months['count'], color=colors)
ax.set_title('Top 10 Months with the Most Netflix Releases')
ax.set_xlabel('Release Month')
ax.set_ylabel('Number of Shows')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

> **Insight:** **November 2019** had the highest number of Netflix releases, followed by January 2020 and December 2019.

### 4.10 Top Content Types in November 2019

```python
nov_2019 = df[(df['Release_Date'].dt.month == 11) & (df['Release_Date'].dt.year == 2019)]
top_types_nov = nov_2019['Type'].value_counts().nlargest(10)

fig, ax = plt.subplots(figsize=(10, 6))
top_types_nov.plot(kind='bar', ax=ax,
    color=['#1f77b4','#ff7f0e','#2ca02c','#d62728','#9467bd','#8c564b','#e377c2','#7f7f7f','#9467bd','#8c564b'])
ax.set_title('Top 10 Content Types Released in November 2019')
ax.set_xlabel('Type')
ax.set_ylabel('Count')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

> **Insight:** **Comedies** were the most released content type during Netflix's peak month of November 2019.
